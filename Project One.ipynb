             4. Data Cleaning
Perfoming Data cleaning involves addressing tasks such as :

Handling missing values

Removing unnecessary characters 

Handling duplicates if any

Converting columns to appropriate datatypes

```python
# Checking out for missing values and handling them
tn_movie_budgets.isnull().sum()
tn_movie_budgets.dropna(inplace=True)
```

```python
#converting columns to appropriate datatypes
tn_movie_budgets['production_budget'] = tn_movie_budgets['production_budget'].str.replace('$', '').str.replace(',', '').astype(int)
tn_movie_budgets['domestic_gross'] = tn_movie_budgets['domestic_gross'].str.replace('$','').str.replace(',', '').astype(int)
tn_movie_budgets['worldwide_gross'] = tn_movie_budgets['worldwide_gross'].str.replace('$', '').str.replace(',', '').astype(int)
print(tn_movie_budgets.head())
```

```python
#Removing the backslash
tn_movie_budgets['movie'] = tn_movie_budgets['movie'].str.rstrip('\\')
print(tn_movie_budgets.head())
```

```python
#Outout the data in tabular format
output = tn_movie_budgets.to_string(index=False)
```

```python
#Removing the white space between release date and movie

tn_movie_budgets['movie'] = tn_movie_budgets['movie'].str.lstrip()
```

```python
#checking for duplicates 
tn_movie_budgets.duplicated().sum()
```

Since there are no duplicates to drop in the tn_move_budgets.csv,we shall proceed with the data transformation process before analysis.

            5.Data Transformation

In this step we are going to convert the relevant columns to appropriate data types.For example the release date column can be converted to a date time data type for easier analysis and manipulation.

```python
#coverting release date column to datetime datatype
tn_movie_budgets['release_date'] = pd.to_datetime(tn_movie_budgets['release_date'])
tn_movie_budgets.dtypes
```

          5. Data Analysis

Here we are going to focus on analyzing the data and and derive insights to make recommendations.We are going to analyze the profitability,Genre performance and Release date.

```python
#Profit calculation by subtracting production budgets from worldwide gross
tn_movie_budgets['profit'] = tn_movie_budgets['worldwide_gross'] - tn_movie_budgets['production_budget']
#Sort profit in descending order
tn_movie_budgets_sorted = tn_movie_budgets.sort_values('profit', ascending=False)
#Get movies with highest profit
print(tn_movie_budgets_sorted)
```

```python
#Outout the data in tabular format
output = tn_movie_budgets.to_string(index=False)
```

```python
#Get the movie with the highest profit
highest_profit_movie = tn_movie_budgets_sorted.iloc[0]
#Printing out the movie
print(highest_profit_movie)
```

From the above cell you can tell that the movie with highest profit is Avatar

```python
#Get the movie with the lowest profit
lowest_profit_movie = tn_movie_budgets_sorted.iloc[-1]
print(lowest_profit_movie)
```

From the above cell you can tell that the movie with the lowest profit is Dark Phoenix

```python
#Exclude non_numeric movie titles
numeric_movies = tn_movie_budgets[pd.to_numeric(tn_movie_budgets['movie'], errors='coerce').notnull()]
#Exploring release dates
#Exploring year and month from the release date
tn_movie_budgets['release_year'] = tn_movie_budgets['release_date'].dt.year
tn_movie_budgets['release_month'] = tn_movie_budgets['release_date'].dt.month
#Group by movie by year and find average worldwide gross and domestic gross
avg_gross_by_year = numeric_movies.groupby('release_year').mean()[['worldwide_gross', 'domestic_gross']]
#Group by month and calculate avg worldwide and domestic gross
avg_gross_by_month = numeric_movies.groupby('release_month').mean()[['worldwide_gross', 'domestic_gross']]
print('avg_gross_by_year:')
print(avg_gross_by_year)
print('\nAvg Gross by Month:')
print(avg_gross_by_month)


```

The above code analyzes the average gross revenue of movies over different years and months thus providing insights into the trends and patterns of movie performance based on release dates.

```python
#Determine the best release year and month based on the average gross revenue
best_year = avg_gross_by_year['worldwide_gross'].idxmax()
best_month = avg_gross_by_month['worldwide_gross'].idxmax()
print("Best Release Year:", best_year)
print("Best Release Month:", best_month)
```

```python
#Exploring  Performance by Genre
#Filter out non-numeric values in the movie column
numeric_movies = tn_movie_budgets[pd.to_numeric(tn_movie_budgets['movie'], errors='coerce').notnull()]
#Split movie column to extract the genre
numeric_movies.loc[:, 'genre'] = numeric_movies['movie'].str.split(':').str[0]
#Group by genre and calculate avg worldwide gross and domestic gross
avg_gross_by_genre = numeric_movies.groupby('genre').mean()[['worldwide_gross', 'domestic_gross']]
print(avg_gross_by_genre)
```

The above analysis provides valuable information for understanding the performance of movies across different genres.
It can be used to inform various decisions in the movie industry,from genre selection to investment strategies.

      6. Data Visualization
      Plays a crucial role in transforming data into meaningful and actionable insights thus facilitating understanding,exploration and design making.

```python
#Bar graph to show movie with the highest profit
plt.figure(figsize=(12, 6))
sns.barplot(x='profit', y='movie', data=tn_movie_budgets_sorted.head(15), color='blue')
plt.xlabel('Profit')
plt.ylabel('Movie')
plt.title('Top 15 Movies with Highest Profit')
#Rotating movie names on x-axis for better readability
plt.xticks(rotation=45)
#Displaying the chart
plt.show()
```

The above visualization represents to 15 movies with the highest profit.
Horizontal value represents the profit value which is the difference between world-wide gross and production budget of each movie.
The vertical axis represents the movie title
The longer bars represents the highest profit while shorter represents comparatively lower profits.

```python
#Plotting average gross by release year

plt.figure(figsize=(12, 6))
sns.lineplot(x='release_year', y='worldwide_gross', data=avg_gross_by_year, label='Worldwide Gross', color='blue')
sns.lineplot(x='release_year', y='domestic_gross', data=avg_gross_by_year, label='Domestic Gross', color='black')
plt.xlabel('Year')
plt.ylabel('Average Gross')
plt.title('Average Gross by Year')
plt.legend()
#Displaying the chart
plt.show()
```

```python
#plotting average gross by release month
plt.figure(figsize=(12, 6))
sns.lineplot(x='release_month', y='worldwide_gross', data=avg_gross_by_month, label='Worldwide Gross', color='blue')
sns.lineplot(x='release_month', y='domestic_gross', data=avg_gross_by_month, label='Domestic Gross', color='black')
plt.xlabel('Month')
plt.ylabel('Average Gross')
plt.title('Average Gross by Month')
plt.legend()
#Displaying the chart
plt.show()
```

The average gross analysis helps us understand the general performance of movies over different release years and months.
It provides insights into their global and domestic earnings
It can be useful for filmmakers,production studios and investors in the film industry to make informes decisions regarding move production,distribution and market targeting

```python
#Plotting average gross by genre
plt.figure(figsize=(12, 6))
sns.scatterplot(x=avg_gross_by_genre.index, y='worldwide_gross', data=avg_gross_by_genre, color='blue', label='Worldwide Gross')
sns.scatterplot(x=avg_gross_by_genre.index, y='domestic_gross', data=avg_gross_by_genre, color='orange', label='Domestic Gross')
plt.xlabel('Genre')
plt.ylabel('Average Gross')
plt.title('Average Gross by Genre')
plt.legend()
#Rotating x-axis label for better readability
plt.xticks(rotation=45)
#Displaying the chart
plt.show()
```

This above shows a scatter plot where each genre has two scatter points representing the average worldwide gross and domestic gross.
The scatter plot allows you to observe the distribution average gross values across genres and identify patterns or outliers in the data
